```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

def plot_power_compare(x,y,power_cone,power_gms,v): # plot function, plot power_cone, power_gms, and diff at a time
    diff = power_cone-power_gms # read power data and compute the diff of the two tests
    
    plt.figure(figsize=(16,4))
    color_map = 'CMRmap'
    
    plt.subplot(131)
    #plt.contour(x,y,power_cone,cmap=plt.cm.brg,linewidths=0.5)
    plt.contourf(x,y, power_cone, 20,cmap=color_map,vmin=v[0], vmax=v[1])
    plt.colorbar()
    xlabel()
    
    plt.subplot(132)
    #plt.contour(x,y,power_gms,cmap=plt.cm.brg,linewidths=0.5)
    plt.contourf(x,y, power_gms, 20,cmap=color_map,vmin=v[0], vmax=v[1])
    plt.colorbar()
    
    plt.subplot(133)
    #plt.contour(x,y,diff,cmap=plt.cm.brg,linewidths=0.5)
    plt.contourf(x,y, diff, 20,cmap=color_map)#,vmin=v[0], vmax=v[1])
    plt.colorbar()
    plt.show()
```

```python
power_list = {"Local-to-boundary":["power_cone_boundary","power_gms_boundary"],
              "Local-to-origin":["power_cone_origin","power_gms_origin"],
              "Small identified set":["power_cone","power_gms"]
             }
v = [0,1]
ngrid = 41
xi, yi = np.meshgrid(
        np.linspace(-2, 2, ngrid).reshape(-1, 1),
        np.linspace(-2, 2, ngrid).reshape(-1, 1),
    )

for name,power in power_list.items():
    power_cone = pd.read_csv(power[0] + '.csv',header=None)
    power_gms = pd.read_csv(power[1] + '.csv',header=None)
    plot_power_compare(xi,yi,power_cone,power_gms,v)
    plt.savefig(name+'.eps', format='eps')
```
