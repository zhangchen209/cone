```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


power_cone = pd.read_csv('power_cone.csv',header=None)
power_gms = pd.read_csv('power_gms.csv',header=None)
diff = power_cone-power_gms # read power data and compute the diff of the two tests

ngrid = 41

xi, yi = np.meshgrid(
        np.linspace(-2, 2, ngrid).reshape(-1, 1),
        np.linspace(-2, 2, ngrid).reshape(-1, 1),
    )

plt.figure(figsize=(16,4))

plt.subplot(131)
plt.contour(xi,yi,power_cone,cmap=plt.cm.brg,linewidths=0.5)
plt.contourf(xi,yi, power_cone, 20,cmap='CMRmap')
plt.colorbar()

plt.subplot(132)
plt.contour(xi,yi,power_gms,cmap=plt.cm.brg,linewidths=0.5)
plt.contourf(xi,yi, power_gms, 20,cmap='CMRmap')
plt.colorbar()

plt.subplot(133)
plt.contour(xi,yi,diff,cmap=plt.cm.brg,linewidths=0.5)
plt.contourf(xi,yi, diff, 20,cmap='CMRmap')
plt.colorbar()

plt.show()
```
